```python
# Path to the folder containing the images
folder_path = "example_jpg"

# List all files in the folder
image_files = os.listdir("example_jpg")

# Filter only JPG files
jpg_files = [file for file in image_files if file.lower().endswith('.jpg')]

# Create a directory for saving prediction images
prediction_folder = "prediction"  # Specify the full path to your prediction folder
os.makedirs(prediction_folder, exist_ok=True)

# Iterate through each JPG file
for index, jpg_file in enumerate(jpg_files):
    # Construct the full path to the JPG file
    jpg_path = os.path.join(folder_path, jpg_file)
    
    # Perform prediction on the JPG image
    prediction_result = model.predict(jpg_path, confidence=20, overlap=30)
    
    # Save the prediction image
    prediction_save_path = os.path.join(prediction_folder, f"prediction_{index + 1}.jpg")
    prediction_result.save(prediction_save_path)
    
    print(f"Processed image {index + 1}: {jpg_file}")

print("Prediction loop completed.")
```
